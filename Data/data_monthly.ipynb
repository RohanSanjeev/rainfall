```python
import pandas as pd 
import numpy as np
```

```python
year_info = pd.read_excel("year_info.xlsx")
year_info.head()
```

```python
non_leap = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
leap = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
```

```python
daily_data = np.load("data_all.npy")
daily_data.shape
```

```python
avg_col = np.zeros((429, 1))
```

```python
for idx, row in year_info.iterrows():

    start = row["start_idx"]

    if row["is_leap"]:
        arr = leap
    else:
        arr = non_leap

    for i in arr:
        end = start + i  
        print(start, end-1, i) 
        avg_col = np.append(avg_col, np.sum(daily_data[:, start:end-1], axis=1).reshape(-1,1)/i, axis=1)
        print(avg_col.shape)
        start = start + i
```

```python
avg_col = avg_col[:, 1:]
avg_col.shape
```

```python
np.save("monthly_avg.npy", avg_col)
```
