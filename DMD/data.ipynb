```python
import pandas as pd
import numpy as np
```

```python
months = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
non_leap_start_index = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
non_leap_end_index = [31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365]

leap_start_index = [1, 32, 61, 92, 122, 153, 183, 214, 245, 275, 306, 336]
leap_end_index = [31, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335, 366]
```

```python
data = pd.read_csv("../eda/NE_available.csv")
data.drop(["Unnamed: 0"], inplace=True, axis=1)
data.head()
```

```python
year = 1901
end_year = 2018
for y in range(year, end_year+1):
    year_data = pd.read_csv(f"../NE_csv/{y}_rain.csv")
    is_leap = True if year_data.shape[1] == 369 else False

    if not is_leap:

        for i,start in enumerate(non_leap_start_index):
            start_col = year_data[f"{start}"]
            for j in range(start+1, non_leap_end_index[i]+1):
                start_col = start_col + year_data[f"{j}"]
            data[f"{y}_{months[i]}"] = start_col/(non_leap_end_index[i]-non_leap_end_index[i]+1)
    else:
        for i,start in enumerate(leap_start_index):
            start_col = year_data[f"{start}"]
            for j in range(start+1, leap_end_index[i]+1):
                start_col = start_col + year_data[f"{j}"]
            data[f"{y}_{months[i]}"] = start_col/(leap_end_index[i]-leap_end_index[i]+1)
```
